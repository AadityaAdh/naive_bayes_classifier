```python

def unique_value_counter(dfr):
  count_dict={}
  for columns in dfr.columns:
    that_column_dict={}
    #nested banaunu xa vanae siddai value halna dido rainxa
    #nested vitra aarko dictionary chai dinu pardo raixa

    for i in  range(dfr.shape[0]):


      my_value=dfr.iloc[i,dfr.columns.get_loc(columns)]
      if my_value not in that_column_dict:
        my_value_count=1
      else:
        my_value_count=that_column_dict[my_value]+1


      that_column_dict[my_value]=my_value_count


    count_dict[columns]=that_column_dict


  return count_dict
```

```python
count_value=unique_value_counter(df_one)
count_value
```

```python
for columns in count_value:
  if columns != "Survived":
    for unique in count_value[columns]:
      count_value[columns][unique] /=count_value["Survived"][1]
```

```python
prob_value_one=count_value
prob_value_one
```

```python
def accuracy_calculator(test_df,prob_value_one,prob_one,prob_value_zero,prob_zero):
  my_columns=test_df.columns[1:]
  correct_pred=0
  total_pred=0

  for i in range(test_df.shape[0]):
    score_zero=prob_zero
    score_one=prob_one
    for j,col in enumerate(my_columns):
      #lets do for survived=1
      if test_df.iloc[i,j+1] in prob_value_one[col]:
        score_one=score_one*prob_value_one[col][test_df.iloc[i,j+1]]
      else:
        score_one=score_one*0



      #lets do for survived=0
      if test_df.iloc[i,j+1] in prob_value_zero[col]:
        score_zero=score_zero*prob_value_zero[col][test_df.iloc[i,j+1]]
      else:
        score_zero=score_zero*0
      #score_zero=score_zero*prob_value_zero[col][test_df.iloc[i,j+1]]

    if(score_one>score_zero):
      belongs=1

    else:
      belongs=0


    if (belongs==test_df.iloc[i,0]):
      correct_pred+=1

    total_pred+=1


  my_accuracy=(correct_pred/total_pred)*100

  return my_accuracy







```

```python
accuracy_calculator(test_df,prob_value_one,prob_one,prob_value_zero,prob_zero)
```
